# Shortest Path Using Dijkstra Algorithm in Python
Problem Description:
Given a weighted graph with non negative edge weights. Find the shortest distance from a source vertex to all other vertices using Dijkstra algorithm

#Theory
Dijkstra algorithm finds shortest distance from one source node to all other nodes in a weighted graph with non negative weights. Dictionary structure stores nodes as keys and neighbor nodes with weights as values.

##Algorithm Steps

1. Represent graph using dictionary.
2. Set distance of every node as infinity.
3. Set distance of source node as 0.
4. Select node with smallest distance among unvisited nodes.
5. Update distance of neighbor nodes.
6. Repeat until all nodes get visited.

```python
def dijkstra(graph, start):

    distances = {}
    visited = []

    for node in graph:
        distances[node] = float('inf')

    distances[start] = 0

    while len(visited) < len(graph):

        min_node = None
        min_distance = float('inf')

        for node in graph:
            if node not in visited and distances[node] < min_distance:
                min_distance = distances[node]
                min_node = node

        visited.append(min_node)

        for neighbor, weight in graph[min_node].items():
            new_distance = distances[min_node] + weight

            if new_distance < distances[neighbor]:
                distances[neighbor] = new_distance

    return distances
```

```python
graph = {
'A': {'B': 4, 'C': 2},
'B': {'A': 4, 'C': 5, 'D': 10},
'C': {'A': 2, 'B': 5, 'D': 3},
'D': {'B': 10, 'C': 3}
}

result = dijkstra(graph, 'A')

for node in result:
    print("Distance from A to", node, "=", result[node])
```

```python
import heapq
def dijkstra(graph, start):

    dist = {node: float('inf') for node in graph}
    prev = {node: None for node in graph}
    dist[start] = 0

    queue = [(0, start)]
    while queue:
        current_dist, current_node = heapq.heappop(queue)

        if current_dist > dist[current_node]:
            continue


        for neighbor, weight in graph[current_node].items():
            new_dist = current_dist + weight
            if new_dist < dist[neighbor]:
                dist[neighbor] = new_dist
                prev[neighbor] = current_node
                heapq.heappush(queue, (new_dist, neighbor))

    return dist, prev

#build path for node so that we can know the path
def build_path(previous, start, end):
    path = []
    current = end
    while current is not None:
        path.insert(0, current)
        current = previous[current]
    return path
# adding node and making it user input
graph = {}
num_nodes = int(input("Enter number of nodes: "))
for i in range(num_nodes):
    node = input(f"Enter name of node {i+1}: ").strip()
    graph[node] = {}

num_edges = int(input("Enter number of edges: "))
for i in range(num_edges):
    u = input("Enter start node: ").strip()
    v = input("Enter end node: ").strip()
    w = int(input("Enter weight: "))
    if u in graph and v in graph:
        graph[u][v] = w
        graph[v][u] = w
    else:
        print(f"Error: one of the nodes {u} or {v} does not exist!")

graph['E'] = graph.get('E', {})
graph['F'] = graph.get('F', {})
graph['E']['F'] = 2
graph['F']['E'] = 2
graph['C']['E'] = 4
graph['E']['C'] = 4
graph['D']['F'] = 11
graph['F']['D'] = 11

sources = list(graph.keys())

for source in sources:
    print(f"\nSource Node: {source}")
    dist, prev = dijkstra(graph, source)
    for node in graph:
        path = build_path(prev, source, node)
        print(f"Shortest path from {source} to {node}: {' -> '.join(path)} "
              f"with distance {dist[node]}")



```
