# 1. Growth Of Function
Time Complexity/Order of Growth defines the amount of time taken by any program with respect to the size of the input. Time Complexity specifies how the program would behave as the order of size of input is increased. So, Time Complexity is just a function of size of its input. We will go through some of basic and most common time complexities such as:

- *Constant Time Complexity O(1) : constant running time*
- *Linear Time Complexity O(n) : linear running time*
- *Logarithmic Time Complexity O(log n) : logarithmic running time*
- *Log-Linear Time Complexity O(n log n) : log-Â­linear running time*
- *Polynomial Time Complexity O(n^c) : polynomial running time (c is a constant)*
- *Exponential Time Complexity O(c^n) : exponential running time (c is a constant being raised to a power based on size of input)*

## Constant time Complexity:O(1)
An algorithm is said to have constant time with order O (1) when it is not dependent on the input size n. Irrespective of the input size n, the runtime will always be the same. See example below

```python
# Prob_1

def add_two_numbers(a, b):
    return a + b

n_values = list(range(1, 101))
time_values = []

for n in n_values:
    add_two_numbers(n, n + 1)
    time_values.append(1)

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("O(1) Constant Time Complexity for Addition")
plt.show()
```

#Linear Time Complexity O(n) : linear running time
- *An algorithm is said to have a linear time complexity when the running time increases linearly with the length of the input. When the function involves checking all the values in input data, with this order O(n).*

```python
def linear_time_sum(numbers):
    total = 0
    for value in numbers:
        total += value
    return total

n_values = list(range(1, 101))
time_values = []

for n in n_values:
    data = list(range(n))
    linear_time_sum(data)
    time_values.append(n)

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("O(n) Linear Time Complexity")
plt.show()
```

#Prob_4 Take Suitable example and write function that runs in logarithic time. plot graph as well similar to above example.

```python
#problem -4 take suitable example and write function that rns in logarithic time
#we have to make grapgh of this also
#let's take example of binary search
import math
import matplotlib.pyplot as plt
def binary_search_steps(n, target):
    low = 0
    high = n - 1
    steps = 0
    while low <= high:
        steps += 1
        mid = (low + high) // 2

        if mid == target:
            return steps
        elif mid < target:
            low = mid + 1
        else:
            high = mid - 1
    return steps
sizes = [10, 20, 40, 80, 160, 320, 640, 1280, 2560, 5120]
steps_list = []
for n in sizes:
    target = n - 1          # worst-case position
    steps = binary_search_steps(n, target)
    steps_list.append(steps)
    print(f"n = {n:4d}, steps = {steps}")
  #let's draw graph of binary search
plt.figure(figsize=(10, 6))
plt.plot(sizes, steps_list, 'bo-', linewidth=2)
plt.xlabel('Input size n')
plt.ylabel('Number of steps')
plt.title('Binary Search Steps (Shows O(log n) Growth)')
plt.grid(True, alpha=0.3)
plt.show()
```

### Polynomial Time Complexity O(n^c) : polynomial running time (c is a constant)

* When the computation time increases as function of N raised to some power, N being the size of input. Such a code has Polynomial Time complexity.




```python
# Prob_6 Take Suitable example and write function that runs in log-linear time. plot the graph as well, similar to above example
import time
import matplotlib.pyplot as plt
import random
#let's do merge sort because we know the time complexity of merge sort
def merge(left, right):
    result = []   # a list which going to store our values here
    i = j = 0      #initialize it with 0 for now and then we c an increment after certain contitions gets true
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    result.extend(left[i:])
    result.extend(right[j:])
    return result
# Merge Sort function
def merge_sort(arr):
    if len(arr) <= 1:
        return arr
    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])
    return merge(left, right)
# Measure runtime for different input sizes
sizes = [100, 200, 400, 800, 1600, 3200]
times = []
for n in sizes:
    arr = [random.randint(0, 10000) for _ in range(n)]
    start = time.time()
    merge_sort(arr)
    end = time.time()
    times.append(end - start)
# let's draw graph
plt.plot(sizes, times, marker='o')
plt.xlabel("Input Size (n)")
plt.ylabel("Time (seconds)")
plt.title("Merge Sort Runtime (O(n log n))")
plt.grid(True)
plt.show()
```

#Exponential Time Complexity O(c^n) : exponential running time (c is a constant being raised to a power based on size of input)

* When the computation time of a code increases as function of X^N, N being the size of input. Such a code has Polynomial Time Complexity.




```python
# Prob_7 Take Suitable example and write function that runs in log-linear time. plot the graph as well, similar to above example
import time
import matplotlib.pyplot as plt
# Simple O(n^2) function: compare all pairs
def polynomial_example(n):
    count = 0
    for i in range(n):
        for j in range(n):
            count += 1   # simple operation count get increment when conditions get true
    return count
# Measure runtime for different input sizes
sizes = [100, 200, 400, 800, 1600]
times = []
for n in sizes:
    start = time.time()
    polynomial_example(n)
    end = time.time()
    times.append(end - start)
# Plotting the growth function
plt.plot(sizes, times, marker='o', color='green')
plt.xlabel("Input Size (n)")
plt.ylabel("Time (seconds)")
plt.title("Polynomial Time Complexity Growth (O(n^2))")
plt.grid(True)
plt.show()
```

```python
# Write a function to perform quick sort and show that it has Logarithmic Time Complexity O(log n)
import random
import time
import math
import matplotlib.pyplot as plt
#Quick Sort Implementation
def partition(arr, low, high):
    pivot = arr[high]           # choose last element as pivot
    i = low - 1                 # index of smaller element
    for j in range(low, high):  # one pass: O(n) work on this segment
        if arr[j] <= pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]
    # place pivot in correct position
    arr[i + 1], arr[high] = arr[high], arr[i + 1]
    return i + 1
def quick_sort(arr, low, high):
    if low < high:
        p = partition(arr, low, high)
        quick_sort(arr, low, p - 1)   # left part
        quick_sort(arr, p + 1, high)  # right part
#Measure Time for Different n
def measure_quicksort_time(n):
    # create random list of size n
    arr = [random.randint(0, n) for _ in range(n)]
    start = time.time()
    quick_sort(arr, 0, len(arr) - 1)
    end = time.time()
    return (end - start) * 1000.0     # to make it in ms
sizes = [100, 200, 400, 800, 1600, 3200, 6400]
times = []
for n in sizes:
    t = measure_quicksort_time(n)
    times.append(t)
    print(f"n = {n:4d}, time = {t:8.3f} ms")
#build n log n Curve for Comparison
nlogn = [n * math.log2(n) for n in sizes]
# scale theoretical curve so its last value matches measured time
scale = times[-1] / nlogn[-1]
nlogn_scaled = [x * scale for x in nlogn]
#plot Graph to Show O(n log n)
plt.figure(figsize=(10, 6))
# measured times (blue)
plt.plot(sizes, times, 'bo-', linewidth=2, label='Measured quicksort time')
#theoretical n log n (red, scaled)
plt.plot(sizes, nlogn_scaled, 'r^-', linewidth=2,
         label='c Â· n logâ n (scaled)')
plt.xlabel('Input size n')
plt.ylabel('Time (ms)')
plt.title('Quicksort Time Complexity: O(n logâ n)')
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```
